```python
# Reload all images
image_5 = mpimg.imread('../../SampleData/000005.jpg')
image_13 = mpimg.imread('../../SampleData/000013.jpg')
image_15 = mpimg.imread('../../SampleData/000015.jpg')
```

```python
plt.imshow(image_5)
```

```python
plt.imshow(image_13)
```

```python
plt.imshow(image_15)
```

```python
# Apply gaussian filter on the image 
image_5_blur = cv2.GaussianBlur(image_5, (5, 5), cv2.BORDER_DEFAULT)
image_13_blur = cv2.GaussianBlur(image_13, (5, 5), cv2.BORDER_DEFAULT)
image_15_blur = cv2.GaussianBlur(image_15, (5, 5), cv2.BORDER_DEFAULT)
```

```python
image_5_gray = cv2.cvtColor(image_5_blur, cv2.COLOR_RGB2GRAY)
image_13_gray = cv2.cvtColor(image_13_blur, cv2.COLOR_RGB2GRAY)
image_15_gray = cv2.cvtColor(image_15_blur, cv2.COLOR_RGB2GRAY)

edges_5 = cv2.Canny(image_5_gray, 50, 150, apertureSize=3)
edges_13 = cv2.Canny(image_13_gray, 50, 150, apertureSize=3)
edges_15 = cv2.Canny(image_15_gray, 50, 150, apertureSize=3)
```

```python
plt.imshow(edges_5, cmap='gray')
```

```python
plt.imshow(edges_13, cmap='gray')
```

```python
plt.imshow(edges_15, cmap='gray')
```

```python
lines_5 = cv2.HoughLines(edges_5, rho=1, theta=np.pi/180, threshold=150)
lines_13 = cv2.HoughLines(edges_13, rho=1, theta=np.pi/180, threshold=150)
lines_15 = cv2.HoughLines(edges_15, rho=1, theta=np.pi/180, threshold=150)
```

```python
print(len(lines))
```

```python
def draw_lines(lines, image):
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*a)
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*a)
        cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 1)
```

```python
draw_lines(lines_5, image_5)
draw_lines(lines_13, image_13)
draw_lines(lines_15, image_15)
```

# Another way, must try

Link: https://forum.opencv.org/t/railway-detection-with-noisy-input/2231

* Get the frame and turn into grayscale
* Apply Gaussian blur
* Apply Sobel operator on x-axis
* Apply Canny edge detector 
* Apply Hough-transform to detect the lines.

```python
# Get the frame an turn ito grayscale
im_1 = mpimg.imread('../../SampleData/1.png')
gray = cv2.cvtColor(im_1, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
print(np.unique(gray))
```

```python
# Apply gaussian blur
# In the paper Power line detection using Hough transform and line tracing techniques
# they suggest using kernel size 5*3 
kernel_size = (5, 3)
blurred = cv2.GaussianBlur(gray, kernel_size, cv2.BORDER_DEFAULT)
plt.imshow(blurred, cmap='gray')
```

```python
# Apply sobel operator on x-axis
sobelx = cv2.Sobel(blurred, cv2.CV_8U, 1, 0, ksize=5)
plt.imshow(sobelx, cmap='gray')
```

```python
sobelx_8bit = (sobelx / np.unique(sobelx)[-1] * 255).astype('uint8')
print(np.unique(sobelx_8bit))
plt.imshow(sobelx_8bit, cmap='gray')
```

```python
wide_range_canny = cv2.Canny(sobelx_8bit, 0, 200)
mid_range_canny = cv2.Canny(sobelx_8bit, 30, 150)
tight_range_canny = cv2.Canny(sobelx_8bit, 240, 250)
```

```python
plt.imshow(wide_range_canny, cmap='gray')
```

```python
plt.imshow(mid_range_canny, cmap='gray')
```

```python
plt.imshow(tight_range_canny, cmap='gray')
```

```python
plt.imsave('Result_1.jpg', tight_range_canny)
```
